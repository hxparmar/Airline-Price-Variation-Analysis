# Q.1. What are the airlines in the dataset, accompanied by their frequencies?

```python
colors=['#003f5c','#444e86','#955196','#dd5182','#ff6e54','#ffa600']

airline_counts=data.groupby('airline').size().reset_index(name='Frequency')
airline_counts
#df.groupby('airline'):
# This groups the dataframe df by the column airline.
# So instead of looking at all rows at once, Pandas will now handle rows grouped by airline name.
# Example: all rows with "Vistara" are one group, all "Indigo" are another, and so on.

# .size():
# After grouping, .size() counts the number of rows in each group.
# So for each airline, it gives how many times that airline appears.
# The result is a Pandas Series with airline names as the index and counts as the values.

# .reset_index(name='Count'):
# By default, the group labels (airline names) are stored as the index in the Series.
# .reset_index() converts the index into a regular column so that we get a proper table.
# The argument name='Count' renames the column of counts to "Count".
```

```python
airline_counts = airline_counts.sort_values(by='Frequency', ascending=False)

plt.barh(airline_counts['airline'], airline_counts['Frequency'], color=colors)#color=["skyblue","orange"])
plt.xlabel("Frequency (Number of Flights)")
plt.ylabel("Airline")
plt.title("Airline Frequency Distribution")
plt.gca().invert_yaxis() #Put the largest value at the top
plt.show()
```

# Q.5. Does ticket price change based on the departure time and arrival time?

```python
mean_departure=data.groupby('departure_time')['price'].mean()
mean_arrival=data.groupby('arrival_time')['price'].mean()
print(mean_departure,"\n\n",mean_arrival)
```

```python
plt.figure(figsize=(10,8))
sns.catplot(x ='price',y ='departure_time',kind ='bar',palette =colors,data =data,hue='departure_time',legend=False)
plt.grid(linestyle='--')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.catplot(x ='price',y ='arrival_time',kind ='bar',palette =colors,data =data,hue='arrival_time',legend=False)
plt.grid(linestyle='--')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.relplot( x = 'departure_time', y = 'price', data = data, col = 'arrival_time', kind = 'line')
plt.show()
```
